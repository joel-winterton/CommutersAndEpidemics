# Commuter Flows and LAD2011 Codes
Save just a matrix, and a corresponding ordered LAD code list (so we can move between datasets given the same geography).

```python
flow_data = pd.read_csv('WF02EW LAD2011.csv', index_col=0, skiprows=9)
flow_data.drop(flow_data.index[range(346, len(flow_data.index))], inplace=True)
flow_data.to_csv('clean/od_matrix.csv', index=False, header=False)
```

```python
flow_data
```

```python
commuter_lads = flow_data.index.values
```

```python
ons_lads = pd.read_csv('LocalAuthorityDistricts2011.csv', index_col=0)
ons_lads
```

```python
commuter_lads[~np.isin(commuter_lads, ons_lads['LAD11NM'])]
```

There's a little bit of naming inconsistency, these exist in the dataset as 'Westminster' and 'Isles of Scilly' so just need to rename them in the ons_lad_names then it's all good.

```python
commuter_lads[np.argwhere(commuter_lads == 'Westminster,City of London')] = 'Westminster'
commuter_lads[np.argwhere(commuter_lads == 'Cornwall,Isles of Scilly')] = 'Isles of Scilly'
```

```python
commuter_lads[np.isin(commuter_lads, ons_lads['LAD11NM'])]
```

```python
filtered_lads = ons_lads[ons_lads['LAD11NM'].isin(commuter_lads)]
filtered_lads = filtered_lads.assign(LAD11NM=pd.Categorical(filtered_lads['LAD11NM'], categories=commuter_lads, ordered=True))
filtered_lads = filtered_lads.sort_values('LAD11NM')
```

```python
np.savetxt('clean/lad_codes.csv', filtered_lads['LAD11CD'].values, delimiter=',', fmt='%s')
```

# Population sizes

```python
population_data = pd.read_csv('QS102EW LAD2011.csv', index_col=0, skiprows=8)
population_data
```

```python
commuter_lads[~np.isin(commuter_lads, population_data.index.values)]
```

ONS commuter data is ground truth. There is some naming inconsistencies between commuter and population datasets, so here I rename them to match and then reorder the population data so that the order is the same as the commuter matrix.


```python
# renaming population data so format is old_name: new_name (e.g. name_in_population_data: name_in_commuter_data)
renamer = {'Folkestone and Hythe': 'Shepway','Vale of Glamorgan':'The Vale of Glamorgan', 'Rhondda Cynon Taff':'Rhondda Cynon Taf'}
```

```python
renamed_pop_index = population_data.index.values
for el in renamer:
    renamed_pop_index[renamed_pop_index == el] = renamer[el]
population_data.set_index(renamed_pop_index, inplace=True)
```

```python
population_data
```

```python
commuter_lads[~np.isin(commuter_lads, population_data.index.values)]
```

```python
sorted_pop_data = population_data.loc[commuter_lads]
np.savetxt('clean/population_counts.csv', sorted_pop_data['2011'].values, fmt='%s')
```
